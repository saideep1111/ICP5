Mean of the house prices

```python
#2 Find the mean of the house prices
mean_price = df['price'].mean()
print(f"Mean house price: ${mean_price:.2f}")
```

Median of the house prices

```python
# 3. Find the median of the house prices
median_price = df['price'].median()
print(f"Median house price: ${median_price:.2f}")
```

Different types of property in the dataset

```python
# 4. Find the different types of property in the dataset
property_types = df['type'].unique()
print("Property types:", property_types)
```

Maximum price of a "condo" in the dataset

```python
# 5. Find the maximum price of a "condo" in the dataset
print(f"The maximum price of a condo is: {df[df['type'] == 'Condo']['price'].max()}")
```

Mean of each type of property in the dataset

```python
# 6. Find the mean of each type of property in the dataset
mean_by_property_type = df.groupby('type')['price'].mean()
print("Mean price by property type:")
print(mean_by_property_type)
```

Count of different property types in dataset

```python
# 7. Find the count of different property types in dataset
property_type_counts = df['type'].value_counts()
print("Count of different property types:")
print(property_type_counts)
```
